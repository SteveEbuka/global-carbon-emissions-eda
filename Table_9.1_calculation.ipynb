```python
import pandas as pd

# Load the dataset (the OWID 'owid-co2-data.csv')
data = pd.read_csv('owid-co2-data.csv')

def calculate_emissions_growth(country_name, start_year=1990, end_year=2021):
    # Filter data for the specific country and years
    country_data = data[data['country'] == country_name]
    
    val_start = country_data[country_data['year'] == start_year]['co2'].values[0]
    val_end = country_data[country_data['year'] == end_year]['co2'].values[0]
    
    # Formula: ((Final - Initial) / Initial) * 100
    growth = ((val_end - val_start) / val_start) * 100
    return growth

# 1. Compute Growth Rates
nigeria_growth = calculate_emissions_growth('Nigeria')
global_growth = calculate_emissions_growth('World')

# 2. Compute Growth Velocity (Relative Acceleration)
growth_velocity = nigeria_growth / global_growth

# 3. Compute Variance
variance = nigeria_growth - global_growth

# Output the results for the table
print(f"Nigeria Growth: {nigeria_growth:.2f}%")
print(f"Global Growth: {global_growth:.2f}%")
print(f"Growth Velocity Index: {growth_velocity:.2f}x")
print(f"Variance: {variance:.2f}%")
```
